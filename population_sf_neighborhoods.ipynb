# Cleaning the population data set to calculate the crime rate per neighborhood

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
# Create a reference the CSV file desired
csv_path ="Resources/COVID-19_Vaccinations_Given_to_SF_Residents_by_Geography.csv"


#Read the csv
pop_df = pd.read_csv(csv_path)


# Print the first five rows of data to the screen

pop_df
```

```python
population_df =pop_df.rename(columns={"id": "Neighborhood", "acs_population":"Population"})
population_df
```

```python
population_to_merge = population_df[["Neighborhood","Population"]]
population_to_merge
```

```python
#New DataFrame with the renamed columns.
population_neighborhood = pd.DataFrame(population_to_merge)
#Save our final DataFrame as a csv.
population_neighborhood.to_csv("population_neighborhood.csv", index=False)
```
