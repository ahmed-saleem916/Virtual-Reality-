```python
import pandas as pd
```

```python
file = "../Resources/Police_Department_Incident_Reports__2018_to_Present.csv"

df = pd.read_csv(file)
df
```

```python
df.info()
```

```python
# removes all columns with null values with the exception of "filed online"
dropped_columns = df.drop(columns=["Invest In Neighborhoods (IIN) Areas", "ESNCAG - Boundary File", "Supervisor District 2012", 
                             "Civic Center Harm Reduction Project Boundary", "Central Market/Tenderloin Boundary Polygon - Updated",
                           "CAD Number", "HSOC Zones as of 2018-06-05", "CNN", "Row ID"])
# since "filed online" only holds a boolean value of "True" if the crime report was filed online by someone other than law enforcement,
# fill_na fills a value of "False" if there is data missing to indicate that the report was filed by law enforcement
fill_na = dropped_columns.fillna("False")
fill_na
```

```python
# data frame showing all the crimes reported by members of the public
filed_online = dropped_columns.dropna(how='any')
filed_online
```

```python
# data frame showing all of the crimes reported by police
police_reports = fill_na.loc[fill_na["Filed Online"] == "False"]
police_reports
```
