```python
import pandas as pd
```

```python
file = "../Resources/Police_Department_Incident_Reports__2018_to_Present.csv"

df = pd.read_csv(file)
df
```

```python
df.info()
```

```python
# removes all columns with null values with the exception of "filed online"
dropped_columns = df.drop(columns=["Invest In Neighborhoods (IIN) Areas", "ESNCAG - Boundary File", "Supervisor District 2012", 
                             "Civic Center Harm Reduction Project Boundary", "Central Market/Tenderloin Boundary Polygon - Updated",
                           "CAD Number", "HSOC Zones as of 2018-06-05", "CNN", "Row ID", "Neighborhoods", "Incident ID", 
                                  "Incident Number", "Current Police Districts", "Report Type Code", "Incident Code", 
                                  "Supervisor District", "Current Supervisor Districts", "Point", "Incident Time"])
# since "filed online" only holds a boolean value of "True" if the crime report was filed online by someone other than law enforcement,
# fill_na fills a value of "False" if there is data missing to indicate that the report was filed by law enforcement
fill_na = dropped_columns.fillna("False")
fill_na
```

```python
# data frame showing all the crimes reported by members of the public
filed_online = dropped_columns.dropna(how='any')
filed_online
```

```python
# data frame showing all of the crimes reported by police
police_reports = fill_na.loc[fill_na["Filed Online"] == "False"]
police_reports
```

```python
# dropping rows where data equals False with the exception of "Filed Online"
police_reports = police_reports[police_reports["Incident Datetime"] != "False"]
police_reports = police_reports[police_reports["Incident Date"] != "False"]
# police_reports = police_reports[police_reports["Incident Time"] != "False"]
police_reports = police_reports[police_reports["Incident Year"] != "False"]
police_reports = police_reports[police_reports["Incident Day of Week"] != "False"]
police_reports = police_reports[police_reports["Report Datetime"] != "False"]
police_reports = police_reports[police_reports["Report Type Description"] != "False"]
police_reports = police_reports[police_reports["Incident Category"] != "False"]
police_reports = police_reports[police_reports["Incident Subcategory"] != "False"]
police_reports = police_reports[police_reports["Incident Description"] != "False"]
police_reports = police_reports[police_reports["Resolution"] != "False"]
police_reports = police_reports[police_reports["Intersection"] != "False"]
police_reports = police_reports[police_reports["Police District"] != "False"]
police_reports = police_reports[police_reports["Analysis Neighborhood"] != "False"]
police_reports = police_reports[police_reports["Latitude"] != "False"]
police_reports = police_reports[police_reports["Longitude"] != "False"]
# police_reports = police_reports[police_reports["Current Supervisor Districts"] != "False"]

```

```python
# cleaning the Incident Category Column
police_reports["Incident Category"] = police_reports["Incident Category"].replace(["Weapons Offence"], "Weapons Offense")
police_reports["Incident Category"] = police_reports["Incident Category"].replace(["Motor Vehicle Theft?"], "Motor Vehicle Theft")
police_reports["Incident Category"] = police_reports["Incident Category"].replace(["Suspicious Occ", "Suspicious"], 
                                                                                  "Suspicious Activity")
police_reports["Incident Category"] = police_reports["Incident Category"].replace(["Human Trafficking (A), Commercial Sex Acts", 
                                                                                  "Human Trafficking (B), Involuntary Servitude", 
                                                                                  "Human Trafficking, Commercial Sex Acts"], 
                                                                                  "Human Trafficking")

# df['column name'] = df['column name'].replace(['old value'], 'new value')
# df['column name'] = df['column name'].replace(['1st old value', '2nd old value', ...], 'new value')
```

```python
# information regarding data types and ensuring no null values
police_reports.info()
```

```python
police_reports.head()
```

```python
# starting to pull the report time from the report datetime column
# converting the Datetime columns to datetime data type
police_reports["Incident Datetime"] = pd.to_datetime(police_reports["Incident Datetime"])
police_reports["Report Datetime"] = pd.to_datetime(police_reports["Report Datetime"])
```

```python
# pulling the time as a separate column for each
police_reports["Incident Time"] = police_reports["Incident Datetime"].dt.time
police_reports["Report Time"] = police_reports["Report Datetime"].dt.time

# police_reports["Report Datetime"].dt.strftime("%p")
```

```python
# pulling the date as a separate column for each
police_reports["Incident Date"] = police_reports["Incident Datetime"].dt.date
police_reports["Report Date"] = police_reports["Report Datetime"].dt.date
```

```python
# removing rows where the Incident Time matches the Report Time
police_duplicates = police_reports.copy()

# mouse_duplicate = merge_mouse_df.loc[merge_mouse_df.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
```

```python
# df[ (df['Name'] == df['Nick_Name'])].index
time_duplicate = police_duplicates[(police_duplicates["Incident Time"] == police_duplicates["Report Time"])].index
time_duplicate
```

```python
# dropping the duplicates by index
# df.drop(index_names, inplace = True)
police_individual = police_duplicates.drop(time_duplicate)
police_individual
```

```python
# checking to see if the incident category is fixed
# crime_counts = police_reports_final.groupby(["Incident Category"]).count()["Incident Datetime"]
# crime_counts
```

```python
police_reports_final = police_individual.reset_index(level=None, drop=True)
police_reports_final.head()
```
