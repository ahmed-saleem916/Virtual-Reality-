```python
incident_report_reduced = incident_report[['Incident Datetime', 'Incident Date','Incident Year','Incident Category','Police District',
       'Analysis Neighborhood', 'Supervisor District','Latitude', 'Longitude',
       'Neighborhoods','Current Supervisor Districts', 'Current Police Districts']]
incident_report_reduced.head()
```

```python
incident_report_dropcolumns =incident_report_reduced.dropna(how="any")
```

```python
incident_report_dropcolumns.count()
```

```python
#Count how many incidents occured in each Analysis Neighborhood

incident_neighborhood_count = incident_report_dropcolumns["Analysis Neighborhood"].value_counts()
#incident_neighborhood_count
```

```python
#Get rid of 2023
incident_report_no2023 = incident_report_dropcolumns.loc[incident_report_dropcolumns["Incident Year"]<2023]
#incident_report_no2023
```

```python
incident_report_no2023["Incident Category"].value_counts()
```

```python
# Set new Index to Incident Category
incident_category_index = incident_report_no2023.set_index("Incident Category")
#incident_category_index.head()
```

```python
sort_category_index = incident_category_index.sort_values("Incident Category")
#sort_category_index
```

```python
# Grab the firts Incident Category by Index, looking for a way to drop the rows
#arson_incident = sort_category_index.iloc[0:1900, 0:15]
#arson_incident.tail()
```

```python
# Another way to drop the rows by index
drop_incident_category= sort_category_index.drop(["Fraud","Lost Property","Miscellaneous Investigation","Traffic Violation Arrest",
                                                 "Other","Other Offenses","Case Closure","Courtesy Report","Civil Sidewalks",
                                                 "Embezzlement","Suicide","Suspicious","Liquor Laws","Human Trafficking (A), Commercial Sex Acts",
                                                  "Human Trafficking, Commercial Sex Acts","Human Trafficking (B), Involuntary Servitude"])

drop_incident_category
```

```python
#Count how many incidents occured in each Analysis Neighborhood
grouped_neighborhood_df = drop_incident_category.groupby(["Analysis Neighborhood","Incident Year","Incident Category" ]).count()

#print(grouped_neighborhood_df)

grouped_neighborhood_df
```
