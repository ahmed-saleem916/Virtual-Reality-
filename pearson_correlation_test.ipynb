```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
# Create a reference the CSV file desired
csv_path ="crime_rate_per_neighborhood.csv"
#csv_path_2 = 

#Read the csv
crime_price = pd.read_csv(csv_path)
#df = pd.read_csv(csv_path_2)

# Print the first five rows of data to the screen

crime_price.head
```

```python
crime_price["Neighborhood"].unique()
```

```python
crime_price_df = crime_price[["Price","Crime Rate"]]
corr = crime_price_df["Price"].corr(crime_price["Crime Rate"], method="pearson")
corr
```

```python
#  Pearson correlation coefficient of -0.3562 between house prices and crime rate means that 
#there is a moderate negative correlation between the two variables. 
#This indicates that as crime rate increases, house prices tend to decrease, and vice versa.
```

```python
neighborhood_corr = crime_price[["Date","Neighborhood","Price","Crime Rate"]]
neighborhood_corr
```

```python
#Calculate the pearson coefficient by neighborhood
#Groupby by neighborhood.
#Do the correlation test between the Price and the Crime Rate with .corr()
#.corr() gives us as a result a matrix. As we are only interested in one result of this matrix we filter with .iloc[]
#[0::2,-1]tells us to filter from 0 to the end of the data frame, the 2 tells us to take only step two and 
#the -1 tells us to filter the column
corr_neighborhood =neighborhood_corr.groupby("Neighborhood")[["Price","Crime Rate"]].corr().iloc[0::2,-1].reset_index()
#corr_neighborhood.columns = ["Neighborhood","Pearson Coeficient"]
corr_neighborhood=corr_neighborhood.rename(columns={"Crime Rate":"Pearson"})
corr_neighborhood=corr_neighborhood[["Neighborhood","Pearson"]]
corr_neighborhood
```
