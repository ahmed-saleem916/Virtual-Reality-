```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
# added later for linregress
from scipy.stats import linregress
```

```python
file = "../Resources/crime_rate_per_neighborhood.csv"

file_violent = "../Resources/violent_crimes_price_rate.csv"
file_nonviolent = "../Resources/nonviolent_crimes_price_rate.csv"

crime_rate = pd.read_csv(file)
violent_crime = pd.read_csv(file_violent)
nonviolent_crime = pd.read_csv(file_nonviolent)
```

```python
crime_rate
```

```python
x_values = crime_rate["Crime Rate"]
y_values = crime_rate["Price"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.ylabel("House Value")
plt.xlabel("Crime Rate")
plt.title("SF Incidents and House Value")
plt.show()
```
